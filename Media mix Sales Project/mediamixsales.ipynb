### Feature Engineering

```python
#introducing lag, rolling mean, expanding mean

df1 = df.copy()
df1
```

```python
#finding the total cost spent by binning the expense categories into 3 major bins - TV, online and offline

df1['tv_spend'] = df1['tv_sponsorships']+df1['tv_cricket']+df1['tv_RON']
df1['offline_spend'] = df1['radio']+df1['NPP']+df1['Magazines']+df1['OOH']
df1['online_spend'] = df1['Social']+df1['Programmatic']+df1['Display_Rest']+df1['Search']
```

```python
#calculating cost of sales ratio - dividing total advertising expenses by sales revenue
#done according to the 3 major bins

df1['tv_salesratio'] = (df1['tv_spend']/df1['sales'])*100
df1['offline_salesratio'] = (df1['offline_spend']/df1['sales'])*100
df1['online_salesratio'] = (df1['online_spend']/df1['sales'])*100
```

```python
#calculating total advertisement expense by adding all costs

df1['tot_ads_cost'] = df1['tv_spend']+df1['offline_spend']+df1['online_spend']
```

```python
df1.columns
```

```python
#Calculating total expense in each quarters
#calculating total sales in each quarters

df1['tot_spend_quarter'] = df1.groupby(['Quarters'])['tot_ads_cost'].transform('sum')
df1['tot_sales_quarter'] = df1.groupby(['Quarters'])['sales'].transform('sum')
```

```python
#calculating individual Return on Investment for all 3 major category expenses
#calculating Total ROI
#calculating ROI for each quarter


df1['tv_ROI'] = (df1['sales']-df1['tv_spend'])/df1['tv_spend']
df1['offline_ROI'] = (df1['sales']-df1['offline_spend'])/df1['offline_spend']
df1['online_ROI'] = (df1['sales']-df1['online_spend'])/df1['online_spend']
df1['tot_ROI'] = (df1['sales']-df1['tot_ads_cost'])/df1['tot_ads_cost']
df1['quarter_ROI'] = (df1['tot_sales_quarter']-df1['tot_spend_quarter'])/df1['tot_spend_quarter']
```

```python
#lag - lag of sales from previous month captured for current month
#rollingmean - with a window size - taking average of all the sales values in the window
#expandingmean - with a current size, successively increasing the size of the window for the next values

df1['lag1'] = df1['sales'].shift(1)
df1['rollingmean'] = df1['sales'].rolling(window = 4).mean()
df1['expandingmean'] = df1['sales'].expanding(2).mean()
df1
```

```python
#dropping null values

df1.dropna(inplace = True)
```

```python
plt.figure(figsize=[25,10])
sns.heatmap(df1.corr(), annot = True, cmap = sns.diverging_palette(220, 10, as_cmap = True))
plt.show()
```

### Model Building

1. 2 models are built, one with the multicollinear variables and the other after removing the multicollinear variables
2. in both models, the results are comparable and RMSE is close for both testing and Training
3. In real time scenario, exclusion of variables depend on the business problem understanding and weightage of the variable given

```python
x = df1_new.drop(['sales'], axis =1)
y = df1_new['sales']

xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size = 0.2, random_state = 1000)

temp_xtrain = sma.add_constant(xtrain)
temp_xtest = sma.add_constant(xtest)

lrmodel = sma.OLS(ytrain,temp_xtrain).fit()
lrmodel.summary()
```

```python
pred_test = lrmodel.predict(temp_xtest)
pred_train = lrmodel.predict(temp_xtrain)
print('Testing R2score=' ,r2_score(ytest, pred_test))
print('Training R2score=' ,r2_score(ytrain, pred_train))
print('Testing RMSE=',np.sqrt(mean_squared_error(ytest, pred_test)))
print('Training RMSE=',np.sqrt(mean_squared_error(ytrain, pred_train)))
```
